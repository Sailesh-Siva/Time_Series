```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('shampoo_sales.csv')
df['Month'] = pd.to_datetime(df['Month'], format='%m-%d')
```

```python
df.set_index('Month',inplace=True)
```

```python
df['Sales'].plot(kind='hist',bins=20,title='Sales')
ax= plt.gca()
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
plt.show()
```

```python
df['Sales'].plot(kind='line',figsize=(8,4),title='Sales')
ax= plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

```python
df['day'] = df.index.day
df['weekday'] = df.index.weekday
```

```python
df['lag_1'] = df['Sales'].shift(1)
df
```

```python
df['lag_2'] = df['Sales'].shift(2)
df
```

```python
df['rolling_mean_5'] = df['Sales'].rolling(5).mean()
df['rolling_std_5'] = df['Sales'].rolling(5).std()
df['rolling_min_5'] = df['Sales'].rolling(5).min()
df['rolling_max_5'] = df['Sales'].rolling(5).max()
df
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

fig, axs = plt.subplots(4, 1,figsize=(12,24)) 

axs[0].xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
axs[0].plot(df['rolling_mean_5'])
axs[0].set_title('Subplot 1') 

axs[1].xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
axs[1].plot(df['rolling_std_5'])
axs[1].set_title('Subplot 2') 

axs[2].xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
axs[2].plot(df['rolling_min_5'])
axs[2].set_title('Subplot 3') 

axs[3].xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
axs[3].plot(df['rolling_max_5'])
axs[3].set_title('Subplot 4') 

plt.tight_layout()
plt.show()
```

```python
df['expanding_mean_5'] = df['Sales'].expanding().mean()
df['expanding_std_5'] = df['Sales'].expanding().std()
df['expanding_min_5'] = df['Sales'].expanding().min()
df['expanding_max_5'] = df['Sales'].expanding().max()
df
```
