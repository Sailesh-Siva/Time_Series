```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
np.random.seed(42)
```

```python
values = np.random.randn(100)
```

```python
values
```

```python
dates = pd.date_range(start = "2023-01-01", end = "2023-04-10",freq = "D")
```

```python
dates
```

```python
df = pd.DataFrame(values,index=dates,columns=['value'])
```

```python
df
```

```python
df['value'].plot(kind='hist',bins=20,title='Value')
ax= plt.gca()
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
plt.show()
```

```python
df['value'].plot(kind='line',figsize=(8,4),title='Value')
ax= plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['value'])
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Synthetic Time Series Dataset")
plt.show()
```

```python
df['year'] = df.index.year
df['month'] = df.index.month
df['day'] = df.index.day
df['weekday'] = df.index.weekday
df
```

```python
df['lag_1'] = df['value'].shift(1)
df
```

```python
df['lag_2'] = df['value'].shift(2)
df
```

```python
df['rolling_mean_5'] = df['value'].rolling(5).mean()
df['rolling_std_5'] = df['value'].rolling(5).std()
df['rolling_min_5'] = df['value'].rolling(5).min()
df['rolling_max_5'] = df['value'].rolling(5).max()
df
```

```python
fig, axs = plt.subplots(4, 1,figsize=(12,24)) 

axs[0].xaxis.set_major_formatter(mdates.DateFormatter('%b'))
axs[0].plot(df['rolling_mean_5'])
axs[0].set_title('Subplot 1') 

axs[1].xaxis.set_major_formatter(mdates.DateFormatter('%b'))
axs[1].plot(df['rolling_std_5'])
axs[1].set_title('Subplot 2') 

axs[2].xaxis.set_major_formatter(mdates.DateFormatter('%b'))
axs[2].plot(df['rolling_min_5'])
axs[2].set_title('Subplot 3') 

axs[3].xaxis.set_major_formatter(mdates.DateFormatter('%b'))
axs[3].plot(df['rolling_max_5'])
axs[3].set_title('Subplot 4') 

plt.tight_layout()
plt.show()
```

```python
df['expanding_mean_5'] = df['value'].expanding().mean()
df['expanding_std_5'] = df['value'].expanding().std()
df['expanding_min_5'] = df['value'].expanding().min()
df['expanding_max_5'] = df['value'].expanding().max()
df
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

fig, axs = plt.subplots(4, 1,figsize=(12,24)) 

axs[0].xaxis.set_major_formatter(mdates.DateFormatter('%b'))
axs[0].plot(df['expanding_mean_5'])
axs[0].set_title('Subplot 1') 

axs[1].xaxis.set_major_formatter(mdates.DateFormatter('%b'))
axs[1].plot(df['expanding_std_5'])
axs[1].set_title('Subplot 2') 

axs[2].xaxis.set_major_formatter(mdates.DateFormatter('%b'))
axs[2].plot(df['expanding_min_5'])
axs[2].set_title('Subplot 3') 

axs[3].xaxis.set_major_formatter(mdates.DateFormatter('%b'))
axs[3].plot(df['expanding_max_5'])
axs[3].set_title('Subplot 4') 

plt.tight_layout()
plt.show()
```
