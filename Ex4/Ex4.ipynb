```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'S:\Coding\Python\TimeSeries\Ex4\daily-min-temperatures.csv')
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'],format = '%Y-%m-%d') 
```

```python
df.set_index('Date',inplace = True)
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(df)
```

```python
df = df.resample('ME').asfreq()
df
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(df)
```

```python
df = df.resample('D')
df = df.interpolate(method = 'quadratic')
df
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(df)
```
