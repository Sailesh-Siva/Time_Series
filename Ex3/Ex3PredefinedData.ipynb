```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
temp_df = pd.read_csv('daily-min-temperatures.csv')
```

```python
temp_df.set_index('Date',inplace = True)
```

```python
temp_df.index = pd.to_datetime(temp_df.index)
```

```python
temp_df.index = temp_df.index.strftime('%d %m') 
```

```python
temp_df.rename(columns = {'Temp':'Temperature'},inplace = True)
```

```python
temp_df = temp_df.iloc[:30]
```

```python
temp_df.head()
```

```python
plt.figure(figsize = (10,5))
plt.plot(temp_df.index,temp_df['Temperature'])
plt.title('Daily Temp in Jan24')
plt.xticks(rotation = 60)
plt.xlabel('Daily')
plt.ylabel('Temp in Celsius')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.scatter(temp_df.index,temp_df['Temperature'])
plt.title('Daily Temp in Jan24')
plt.xticks(rotation = 60)
plt.xlabel('Daily')
plt.ylabel('Temp in Celsius')
plt.show()
```

```python
plt.figure(figsize = (10,5))
pd.plotting.autocorrelation_plot(temp_df['Temperature'])
plt.title('Autocorrelation Plot')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.hist(temp_df['Temperature'],bins = 10)
plt.title('Daily Temp in Jan24')
plt.xlabel('Temp in Celsius')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(temp_df['Temperature'],fill = True)
plt.title('Density plot for Daily Temp in Jan24')
plt.xlabel('Temperature')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x = temp_df['Temperature'])
plt.title('Box and Whisker plt for Daily Temp in Jan 24')
plt.xlabel('Temperature')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(temp_df)
plt.title('Box and Whisker plt for Daily Temp in Jan 24')
plt.show()
```
