```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
date_range = pd.date_range(start = '1/1/2024', end = '1/31/2024',freq = 'D')
date_range
```

```python
temp_data = np.random.normal(loc = 10, scale = 5,size = len(date_range))
temp_data
```

```python
temp_df = pd.DataFrame({'Date':pd.to_datetime(date_range),'Temperature':temp_data})
temp_df
```

```python
temp_df.set_index('Date',inplace = True)
```

```python
temp_df.index = temp_df.index.strftime('%d') 
```

```python
temp_df.head()
```

```python
plt.figure(figsize = (10,5))
plt.plot(temp_df.index,temp_df['Temperature'])
plt.title('Daily Temp in Jan24')
plt.xlabel('Daily')
plt.ylabel('Temp in Celsius')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.scatter(temp_df.index,temp_df['Temperature'])
plt.title('Daily Temp in Jan24')
plt.xlabel('Daily')
plt.ylabel('Temp in Celsius')
plt.show()
```

```python
plt.figure(figsize = (10,5))
pd.plotting.autocorrelation_plot(temp_df['Temperature'])
plt.title('Autocorrelation Plot')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.hist(temp_df['Temperature'],bins = 10)
plt.title('Daily Temp in Jan24')
plt.xlabel('Temp in Celsius')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(temp_df['Temperature'],fill = True)
plt.title('Density plot for Daily Temp in Jan24')
plt.xlabel('Temperature')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x = temp_df['Temperature'])
plt.title('Box and Whisker plt for Daily Temp in Jan 24')
plt.xlabel('Temperature')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(temp_df)
plt.title('Box and Whisker plt for Daily Temp in Jan 24')
plt.show()
```
